```python
# --- 4. Derived / transformed columns ---
# log transforms (for skewed variables)
develop["log_GDPpc"] = np.log10(develop["GDP_per_Capita"].replace(0, np.nan))
develop["log_LCC"] = np.log10(develop["LCC"].replace(0, np.nan))
develop["log_Languages"] = np.log10(develop["Languages_spoken"].replace(0, np.nan))
develop["log_Pop_millions"] = np.log10(develop["Population_millions"].replace(0, np.nan))
```

```python
# Preparing features for clustering
# We'll cluster on standardized log_LCC, log_GDPpc, HDI
cluster_df = develop[["Country", "log_LCC", "log_GDPpc", "HDI"]].dropna()
X = cluster_df[["log_LCC", "log_GDPpc", "HDI"]].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(1,7)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=20)
    km.fit(X_scaled)
    inertia.append(km.inertia_)
# plot elbow
plt.figure(figsize=(6,4))
plt.plot(K_range, inertia, marker='o')
plt.xlabel("k (number of clusters)")
plt.ylabel("Inertia")
plt.title("Elbow Method for KMeans")
plt.grid(True, ls="--", alpha=0.5)
plt.savefig(os.path.join(RESULTS_DIR, "elbow_plot.png"), dpi=300)
plt.close()
print(f"Saved elbow plot to {os.path.join(RESULTS_DIR, 'elbow_plot.png')}")
```

```python
# Choose k = 2 or 3 based on elbow; default to 2 for developed vs developing
K_CHOSEN = 2
kmeans = KMeans(n_clusters=K_CHOSEN, random_state=42, n_init=50)
cluster_labels = kmeans.fit_predict(X_scaled)
cluster_df["cluster"] = cluster_labels
df = develop.merge(cluster_df[["Country","cluster"]], on="Country", how="left")
```

```python
# Cluster centers in original space (inverse transform) for interpretation 
centers_scaled = kmeans.cluster_centers_
centers_unscaled = scaler.inverse_transform(centers_scaled)
centers_df = pd.DataFrame(centers_unscaled, columns=["log_LCC","log_GDPpc","HDI"])
centers_df["LCC"] = 10**centers_df["log_LCC"]
centers_df["GDP_per_Capita"] = 10**centers_df["log_GDPpc"]
centers_df["cluster"] = centers_df.index
```

```python
# Feature importance: predict log_GDPpc using RandomForest 
# Prepare feature set: Languages_spoken, Population_millions, LCC, HDI
feat_cols = ["Languages_spoken", "Population_millions", "LCC", "HDI"]
df_feat = df[feat_cols + ["log_GDPpc"]].dropna()
X_feat = df_feat[feat_cols].values
y_feat = df_feat["log_GDPpc"].values

X_train, X_test, y_train, y_test = train_test_split(X_feat, y_feat, test_size=0.2, random_state=42)
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
r2 = r2_score(y_test, y_pred)
importances = rf.feature_importances_
feat_imp = pd.DataFrame({"feature": feat_cols, "importance": importances}).sort_values("importance", ascending=False)
print(f"RandomForest R2 on test set: {r2:.3f}")
print("Feature importances:\n", feat_imp)
```

```python
# Plot feature importance
plt.figure(figsize=(6,4))
sns.barplot(data=feat_imp, x="importance", y="feature", palette="muted")
plt.title("Feature importance (predicting log GDP per capita) - RandomForest")
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "feature_importance.png"), dpi=300)
plt.close()
print(f"Saved feature importance plot to {os.path.join(RESULTS_DIR, 'feature_importance.png')}")
```
